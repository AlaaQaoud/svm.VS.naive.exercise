```python
# قائمة الكلمات التي ليس لها فائدة
stop_words = set([
    "if", "and", "or", "the", "is", "to", "for", "in", "of", "was", "a", "on", "at", "it", "this", "that", "with", "by", "as", "an", "be"
])
```

```python
# دالة لتنظيف النصوص
def clean_text(text):
    # تحويل النص إلى أحرف صغيرة
    text = text.lower()
    # إزالة الرموز الخاصة والأرقام
    text = re.sub(r'[^a-z\s]', '', text)
    # إزالة الكلمات التي ليست ذات فائدة
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text
```

```python
import re
# تنظيف النصوص في مجموعة البيانات
reviews_cleaned = [(clean_text(review), label) for review, label in reviews]
df = pd.DataFrame(reviews_cleaned, columns=["review", "label"])
df
```

```python
#  تحويل النصوص إلى بيانات رقمية باستخدام TF-IDF
# TF -> Equation of word weight
#  IDF -> Frequency of word in all comments
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["review"])
y = df["label"]
```

```python
# تقسيم البيانات إلى تدريب واختبار
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24)
```

```python
#  تدريب النماذج
# نموذج SVM
from sklearn.model_selection import GridSearchCV

#  تحديد المعلمات التي سيتم البحث عنها
param_grid = {
    'C': [1, 10, 100],               
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],  
    'degree': [2, 3]               
}

# استخدام GridSearchCV للبحث عن أفضل المعلمات
grid_search = GridSearchCV(SVC(), param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

# عرض أفضل المعلمات
print("best parameters SVM:", grid_search.best_params_)
```

```python
# استخدام أفضل معلمات لتدريب النموذج
best_svm=SVC(kernel="rbf", degree=2, C=10)
# تدريب النموذج باستخدام أفضل المعلمات
best_svm.fit(X_train, y_train)
```

```python
# نموذج Naïve Bayes
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
```

```python
#  اختبار النماذج
test_reviews = [
    "What a fantastic experience! The staff was friendly, and the food was absolutely delicious.",  
    "The wait was way too long, and the food was cold when it finally arrived."
]
test_features = vectorizer.transform(test_reviews)
```

```python
# التنبؤ
svm_predictions = best_svm.predict(test_features)
nb_predictions = nb_model.predict(test_features)

# عرض النتائج
print("SVM Predictions:", svm_predictions)
print("Naïve Bayes Predictions:", nb_predictions)
```

Yes, both models agree on the classification

```python
# تقييم النماذج
y_pred_svm = best_svm.predict(X_test)
y_pred_nb = nb_model.predict(X_test)


print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print("Naïve Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))
```
